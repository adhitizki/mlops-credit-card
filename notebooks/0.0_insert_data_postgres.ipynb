This file is not the main part of the `credit_card` project, just to demonstrate how the data were ingested from the database. And explain about the data will be continously added to the database when the users write a review about products.

So, the variable will be hard coded here as variable:

```python
PG_USER = os.environ['POSTGRES_USER']
PG_PASS = os.environ['POSTGRES_PASSWORD']
PG_ADDR = "localhost"
DB = "credit_card"
```

```python
import psycopg2 
from sqlalchemy import create_engine 

conn_string = f"postgresql://{PG_USER}:{PG_PASS}@{PG_ADDR}/{DB}"

db = create_engine(conn_string)
conn = db.connect()
```

Insert items to the database.

```python
df = pd.read_csv("notebooks/data/creditcard_2023.csv")
```

```python
df.head()
```

```python
TABLE_ITEMS = "data_features"
```

```python
df.to_sql( 
    f"{TABLE_ITEMS}", 
    con=conn, 
    if_exists="replace", 
    index=False
)
```
